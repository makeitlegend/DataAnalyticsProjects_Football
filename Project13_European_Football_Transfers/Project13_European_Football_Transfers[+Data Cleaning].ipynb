```python
#SEMICOLON-SEPARATED FACT TABLES
df_clubs_in_leagues = pd.read_csv("/content/clubs_in_leagues.csv", sep=";")
df_coaches_in_clubs = pd.read_csv("/content/coaches_in_clubs.csv", sep=";")
df_nat_stats_of_players = pd.read_csv("/content/nat_stats_of_players.csv", sep=";")
df_players_in_clubs = pd.read_csv("/content/players_in_clubs.csv", sep=";")
df_pos_stats_of_players = pd.read_csv("/content/pos_stats_of_players.csv", sep=";")
df_stats_of_players = pd.read_csv("/content/stats_of_players.csv", sep=";")
df_transfers = pd.read_csv("/content/transfers.csv", sep=";")

#CLEAN TSV DICTIONARIES
df_dict_countries = pd.read_csv(
    "/content/dict_countries.csv",
    sep=";",
    engine="python",
    encoding="utf-8"
)

df_dict_leagues = pd.read_csv(
    "/content/dict_leagues.csv",
    sep=";",
    engine="python",
    encoding="utf-8"
)

df_dict_stadiums = pd.read_csv(
    "/content/dict_stadiums.csv",
    sep=";",
    engine="python",
    encoding="utf-8"
)

#DIRTY TSV DICTIONARIES
df_dict_clubs = pd.read_csv(
    "/content/dict_clubs.csv",
    sep="\t",
    header=None,
    names=["club"],
    engine="python",
    encoding="utf-8",
    on_bad_lines="skip"
)

df_dict_players = pd.read_csv(
    "/content/dict_players.csv",
    sep="\t",
    header=None,
    names=["player"],
    engine="python",
    encoding="utf-8",
    on_bad_lines="skip"
)
```
